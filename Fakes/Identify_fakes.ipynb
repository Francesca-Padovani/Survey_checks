```python
#function that open the file and convert categorical variable to ordinal numbers and return list of odd and even answers for the variable PERS

def check_answers_PERS(doc, variable):
    df = pd.read_spss(doc)
    
    list_of_even = []
    list_of_odd = []
    
    for col in df.columns:
        if variable in col:
            locus = col.rfind('_')
            number = int(col[locus+1:])
            panelista = df['panel'].tolist()
            
            to_change = df[col].tolist()
            changed = []
                
            for ele in to_change:
                
                if ele == "Per niente d'accordo":
                    
                    m = 1
                    changed.append(m)

                elif ele == "Non molto d'accordo":
                    m = 2
                    changed.append(m)

                elif ele == "Abbastanza d'accordo":
                    m = 3
                    changed.append(m)

                elif ele == "Molto d'accordo":
                    m = 4
                    changed.append(m)
                
                else: 
                    m = 0
                    changed.append(m)
                        
            if number % 2 == 0:
                list_of_even.append(changed)
                
            else:
                list_of_odd.append(changed)
            
    triplette = []
    for j in range(len(list_of_odd[0])):
        risp_disp = []
        risp_pari = []

        for i in range(len(list_of_odd)):

            risp_disp.append(list_of_odd[i][j])
            risp_pari.append(list_of_even[i][j])

        triplette.append((f'Utente_{j}', panelista[j], risp_disp, risp_pari))
    
    return triplette
        
```

```python
#apply the function defined above 
pers = check_answers_PERS(file, 'PERS')

#calculate how many questionable answers are present here, control also for the panelist
spot_liers = []
for ele in pers:
    count = 0
    for i in range(5):
        if ele[2][i] == 4 and ele[3][i] == 4:
            count += 1
        
        elif ele[2][i] == 1 and ele[3][i] == 1:
            count += 1
        
    spot_liers.append((ele[0], ele[1], count))
    

print(f'La quantità di utenti che hanno risposte contrastanti è: ', sum([1 for ele in spot_liers if ele[2] >=2]))  

#print them out
for record in spot_liers:
    if record[2] >=2:
        print(record)
```

```python
def check_answers_2(doc, variable, soglia = 5):
    
    df = pd.read_spss(doc)
    
    if variable == 'INDCOL':
        data = df[[f"{variable}_1",f"{variable}_2",f"{variable}_3", f"{variable}_4", f"{variable}_5", f"{variable}_6", f"{variable}_7", 'panel']]
        
        for col in data.columns: 
            if not col == 'panel':
                data[col] = data[col].replace(['Molto', 'Abbastanza', 'Poco', 'Per niente'], [4, 3, 2,1])
        
    elif variable == 'CONS':
        data = df[[f"{variable}_1",f"{variable}_2",f"{variable}_3", f"{variable}_4", f"{variable}_5", f"{variable}_6", 'panel']]
        
        for col in data.columns:
            if not col == 'panel':
                data[col] = data[col].replace(["Molto d'accordo", "Abbastanza d'accordo", "Non molto d'accordo", "Per niente d'accordo"], [4, 3, 2,1])
        
    
        
    amount_red_flags = []
    
    # loop through the rows using iterrows()
    for index, row in data.iterrows():
        c = 0
        lista_risposte = row.tolist()
        count_4 = sum([1 for ele in lista_risposte if ele == 4])
        count_1 = sum([1 for ele in lista_risposte if ele == 1])
        
        
        if count_4 >=soglia:
            c += 1
            
        elif count_1 >=soglia:
            c += 1
        
        amount_red_flags.append((f'Utente_{index}', row['panel'], c))
    
    total = sum([1 for elem in amount_red_flags if elem[2] >= 1])
    which = [(elem[0],elem[1], elem[2]) for elem in amount_red_flags if elem[2] >= 1]
    
    return total, which
        
```

```python
#individualismo vs collettivismo 
indcol = check_answers_2(file, 'INDCOL')
print(f"La quantità di persone dalle risposte dubbie è: {indcol[0]}")
print()
for item in indcol[1]:
    print(item)
```
